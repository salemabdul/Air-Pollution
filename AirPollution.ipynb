```python
# Importing the matplotlib library:
import matplotlib.pyplot as plt
```

```python
# Importing the seaborn package:
import seaborn as sns
```

```python
# Plotting the pm10 concentration levels for each of the cities in Sudan on a bar plot:
barplot = sns.barplot(x="City", y="1999 pm10 concentration", data=Sudan_pm10)

# Rotating the values on the x-axis with a For-loop:
for item in barplot.get_xticklabels():
    item.set_rotation(70)

# Adding a title for the bar chart:
plt.title("pm10 Concentration levels in the Cities in Sudan")
```

- As we can see from the bar chart above, we can clearly see which cities have the highest and lowest pm10 concentration levels.

```python
# Selecting only the rows, that have the name Belarus as a country:
Belarus_pm10 = City_PM_sorted_pm10.loc[City_PM_sorted_pm10.country == "Belarus"]
```

```python
# Printing out the Belarus DataFrame:
Belarus_pm10
```

However, we can say that MINSK has the highest pm10 concentration level and Solegorsk has the lowest pm10 concentration level. Just like with Sudan, we can plot this in a bar chart.

```python
# Plotting the pm10 concentration levels for each of the cities in Sudan on a bar plot:
barplot_2 = sns.barplot(x="City", y="1999 pm10 concentration", data=Belarus_pm10)

# Rotating the values on the x-axis with a For-loop:
for item in barplot_2.get_xticklabels():
    item.set_rotation(70)

# Adding a title for the bar chart:
plt.title("pm10 Concentration levels in the Cities in Belarus")
```
